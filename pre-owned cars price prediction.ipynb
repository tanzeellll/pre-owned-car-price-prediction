```python
y_train = ds_train.pop('selling_price')
X_train = ds_train
```

```python
X_train.head()
```

### Building the model

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
rfe = RFE(lm, n_features_to_select=10)
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns, rfe.support_, rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

```python
X_train_rfe = X_train[col]
```

```python
import statsmodels.api as sm
```

```python
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
lm = sm.OLS(y_train, X_train_rfe).fit()
print (lm.summary())
```

```python
X_train_new = X_train_rfe.drop('owner_Test Drive Car', axis=1)
```

```python
X_train_lm = sm.add_constant(X_train_new)
```

```python
lm = sm.OLS(y_train, X_train_lm).fit()
print (lm.summary())
```

### Checking for VIF

```python
X_train_new = X_train_new.drop('const', axis = 1)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor


vif = pd.DataFrame()
X = X_train_new
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range (X.shape[1])]
vif = vif.sort_values(by = 'VIF', ascending=False)
vif
```

```python
X_train_new_1 = X_train_new.drop('transmission_Manual', axis=1)
```

```python
X_train_lm_1 = sm.add_constant(X_train_new_1)
```

```python
lm = sm.OLS(y_train, X_train_lm_1).fit()
print (lm.summary())
```

```python
vif = pd.DataFrame()
X = X_train_new_1
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range (X.shape[1])]
vif = vif.sort_values(by = 'VIF', ascending=False)
vif
```

### Rasidual Analysis

```python
y_train_price = lm.predict(X_train_lm_1)
```

```python
sns.distplot(y_train-y_train_price, bins = 20)
plt.show()
```
